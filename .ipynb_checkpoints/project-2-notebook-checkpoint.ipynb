```python
# Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
#from config import pg_pass
#from config import pg_user
import psycopg2
```

```python
# create house df from csv
house_csv = "resources/melb_data.csv"
house_df = pd.read_csv(house_csv)
house_df.head()
```

```python
# create venues df from csv
venus_csv = "resources/Melbourne_venues.csv"
venus_df = pd.read_csv(venus_csv)
venus_df.head()
```

```python
# transform house table
house_cols = ["Suburb", "Address", "Price", "CouncilArea"]
melbourne_housing = house_df[house_cols].copy()

# check new list
melbourne_housing.head()
```

```python
# transform venue table
venue_cols = ["Suburb", "Venue Category", "Venue"]
melbourne_venue = venus_df[venue_cols].copy()

# check new list
melbourne_venue.head()
```

```python
# create engine for postgresql
#connection_string = (f'{pg_user}:{pg_pass}@localhost:5432/Melbourne_data_db')
#engine = create_engine(f'postgresql://{connection_string}')
rds_connection_string = "postgres:4167@localhost:5432/Melbourne_data_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# check table names
engine.table_names()
```

```python
melbourne_venue.to_sql(name='melb_venue', con=engine, if_exists='append', index=False)
```

```python
melbourne_housing.to_sql(name='melbourne_housing', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from melb_venue', con=engine).head()
```

```python
pd.read_sql_query('select * from melbourne_housing', con=engine).head()
```
